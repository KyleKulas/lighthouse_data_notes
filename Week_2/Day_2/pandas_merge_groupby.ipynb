# Pandas Merge and Group By

```python
import pandas as pd
import numpy as np
```

## Merge

```python
df = pd.DataFrame(np.random.randn(10,4))
df
```

```python
# break into pieces
pieces = [df[:3], df[3:7], df[7:]]
pieces
```

```python
pd.concat(pieces)
```

## Merge

```python
left = pd.DataFrame({'key': ['foo', 'bar', 'qux'], 'lval': [1,2, 4]})
left
```

```python
right = pd.DataFrame({'key': ['foo', 'bar', 'foo', 'baz'], 'rval': [4, 5, 6, 7]})
right
```

```python
pd.merge(left, right, on='key', how='left')
```

## Grouping

-splitting the data into groups on some criteria

-Applying a function to each group independently

-Combining the results into a data structure

```python
df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
                    'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
                    'C': np.random.randn(8),
                    'D': np.random.randn(8)})
df
```

```python
df.groupby('A').sum()
```

```python
df.groupby(['A','B']).sum()
```

```python
df.groupby('A').agg({'C': np.sum, 'D': np.max})
```

## Stack

```python
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
                         'foo', 'foo', 'qux', 'qux'],
                        ['one', 'two', 'one', 'two',
                         'one', 'two', 'one', 'two']]))
index = pd.MultiIndex.from_tuples(tuples, names=['first','second'])

df = pd.DataFrame(np.random.randn(8,2), index=index, columns=['A','B'])
df
```

```python
df2 = df[:4]

df2
```

```python
stacked = df2.stack()
stacked
```

```python
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

## Pivot Tables

```python
df = pd.DataFrame({'A': ['one', 'one', 'two', 'three'] * 3,
                   'B': ['A', 'B', 'C'] * 4,
                   'C': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                   'D': np.random.randn(12),
                   'E': np.random.randn(12)})
df
```

```python
pd.pivot_table(df, values='D', index=['A','B'], columns=['C'])
```

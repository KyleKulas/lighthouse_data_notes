```python
import pandas as pd
```

```python
my_dict = {'name': ['a', 'b', 'c', 'd', 'e', 'f', 'g'],
           'age': [20, 27, 35, 55, 18, 21, 35],
           'designation': ["VP", "CEO", "CFO", "VP", "VP", "CEO", "MD"]}

df = pd.DataFrame(my_dict)
df
```

```python
df.to_csv('csv_example.csv', index=False)
```

```python
df_csv = pd.read_csv('csv_example.csv')
df_csv
```

```python
df_csv = pd.read_csv('csv_example.csv', header = [0, 1, 2])
df_csv
```

```python
df_csv = pd.read_csv('csv_example.csv', header = 5)
df_csv
```

```python
df_csv = pd.read_csv('csv_example.csv', header=[1,2,5])
df_csv
```

```python
df_csv = pd.read_csv('csv_example.csv', names = ['a', 'b', 'c'])
df_csv
```

```python
df_csv = pd.read_csv('csv_example.csv', header=1, names=['a', 'b', 'c'])
df_csv
```

## CSV to (anything) separated value

```python
df.to_csv('csv_example.csv', index=False, sep=':')
```

```python
df_csv = pd.read_csv('csv_example.csv', sep=':')
df_csv
```

Set index:

```python
df_csv.set_index('age')
```

Or the same thing can be done when reading file

```python
df_csv = pd.read_csv('csv_example.csv', sep=':', index_col=1)
df_csv
```

To only load a certain number of rows:

```python
df_csv = pd.read_csv('csv_example.csv', sep=':', nrows=3)
df_csv
```

## skipping empty lines in csv files

by default `read_csv(...)` will skip blank lines. To read them:

```python
df_csv = pd.read_csv('csv_example.csv', skip_blank_lines=False, sep=':')
df_csv
```

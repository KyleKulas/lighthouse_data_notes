### Task 2
Test the hypothesis that the offensive productions stats of the Cleveland Cavaliers and Golden State Warriors (the teams that met in the finals that year) were from the same distribution in the 2015/2016 season.

Offensive production refers to two variables: **PTS (Points)** and **FG_PCT (Field Goal Percentage)**. We will need to do two separate hypothesis tests, one for each variable.

```python
cavaliers_points = nba_22015[nba_22015['TEAM_NAME'] == 'Cleveland Cavaliers']['PTS']
warriors_points = nba_22015[nba_22015['TEAM_NAME'] == 'Golden State Warriors']['PTS']

cavaliers_fgp = nba_22015[nba_22015['TEAM_NAME'] == 'Cleveland Cavaliers']['FG_PCT']
warriors_fgp = nba_22015[nba_22015['TEAM_NAME'] == 'Golden State Warriors']['FG_PCT']
```

```python
independent_sample_ttest_points = st.ttest_ind(cavaliers_points, warriors_points)

independent_sample_ttest_fgp = st.ttest_ind(cavaliers_fgp, warriors_fgp)
```

- **Coach Blatt was fired on 24th of Jan, 2016**. 

Use the data from seasons 2014/2015 and 2015/2016 - those are years when Cleveland was coached by Blatt. 

**We have two possible solutions to try here:**
- Take the same amount of games from before and after and try t-test.
- Take all the games from before and after and look for the right test to compare two samples with different sizes. (You will need to go through the scipy documentation or google to figure out what kind of test is required.)

```python
nba_22015['GAME_DATE'] = pd.to_datetime(nba_22015['GAME_DATE'])
nba_22015.head()
```

```python
# get all the cleveland cavaliers games after blatt was fired on 2016-01-04
after_blatt = nba_22015[nba_22015['GAME_DATE'] > pd.Timestamp('2016-01-24')][nba_22015['TEAM_NAME'] == 'Cleveland Cavaliers']
after_blatt.sort_values('GAME_DATE')
```

```python
# get 45 games before firing blatt

# concat previous season
concated = pd.concat([nba_22015,nba_22014])

#convert game date column to time stamp
concated['GAME_DATE'] = pd.to_datetime(concated['GAME_DATE'])

# get 45 games before firing blatt
with_blatt = concated[concated['GAME_DATE'] < '2016-01-24'][concated['TEAM_NAME'] == 'Cleveland Cavaliers']
with_blatt = with_blatt.sort_values('GAME_DATE', ascending=False).head(40)
```

```python
independent_sample_ttest_blatt = st.ttest_ind(after_blatt['PTS'], with_blatt['PTS'])
```
